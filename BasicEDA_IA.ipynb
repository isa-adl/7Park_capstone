```python
#season trends 

def month_to_season(month):
    if month <= 5:
        month = 'spring'
    elif month <= 8:
        month = 'summer'
    elif month <= 11:
        month = 'fall'
    else:
        month = 'winter'
    return(month)

ZRI_f['Season'] = ZRI_f.Month.apply(month_to_season)
```

```python
season_rent = ZRI_f.groupby(['Season']).agg({"value": 'median'})
```
