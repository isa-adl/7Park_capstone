```python
starbucks['Opened'] = pd.to_datetime(starbucks["Opened"])
```

```python
starbucks['Name1']= starbucks['Name'].apply(lambda s : s.split('and')[0])
```

```python
starbucks['Name2']= starbucks['Name'].apply(lambda s : s.split('and')[1])
```

```python
starbucks['Market'] = starbucks['Market'].apply(lambda s : s.split('\\')[0])
```

```python
starbucks['Address1'] = starbucks['Name1'] + ', ' + starbucks['City'] + ', ' + starbucks['Market']
```

```python
starbucks['Address2'] = starbucks['Name'].apply(lambda s : s.split('&')[0])
```
